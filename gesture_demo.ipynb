## Read the primary-hand result

Re-run these cells while the camera is running.

```python
print("gesture   :", w.gesture)
print("confidence:", round(w.confidence, 3))
print("handedness:", w.handedness)
print("num hands :", w.num_hands)
print("fps       :", round(w.fps, 1))
```

```python
# Individual finger / landmark positions ([x, y, z], normalized 0..1)
print("index_finger_tip:", w.index_finger_tip)
print("thumb_tip       :", w.thumb_tip)
print("wrist           :", w.wrist)
```

```python







# All 21 landmarks at once
for name, pos in w.finger_positions().items():
    pretty = [round(c, 3) for c in pos] if pos else "—"
    print(f"{name:18s} {pretty}")
```

## Observe a landmark live

The callback fires every time the index fingertip position updates
(throttled by `w.sync_interval_ms`, default 100 ms).

```python
import ipywidgets as widgets
from IPython.display import display

out = widgets.Label(value="index_finger_tip: (waiting…)")

def _on_tip(change):
    p = change["new"]
    out.value = (
        f"index_finger_tip: x={p[0]:.3f} y={p[1]:.3f} z={p[2]:.3f}"
        if p else "index_finger_tip: (no hand)"
    )

w.observe(_on_tip, "index_finger_tip")
display(out)
```
